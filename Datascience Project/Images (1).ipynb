```python
import numpy as np
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.vgg16 import preprocess_input
```

```python
img=image.load_img('lena.png',target_size=(224,224))
```

```python
#convert to array
img_array=image.img_to_array(img)
```

```python
img_array=np.expand_dims(img_array,axis=0)
```

```python
img_array=preprocess_input(img_array)
```

```python
#image augmentation ---- it helps you to generate different variation of the image
#generate different variation of image  using rotation,flipping,zooming etc
```

```python
#tensflow dediacted to work with neural networks
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
datagen=ImageDataGenerator(rotation_range=20,
                           width_shift_range=0.20,
                           height_shift_range=0.2,
                           zoom_range=0.2,
                           horizontal_flip=True,
                           fill_mode='nearest')
```

```python
aug_item=datagen.flow(img_array)
```

```python
aug_images=[next(aug_item)[0].astype(np.uint8) for i in range(5)]
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
for i,img in enumerate(aug_images):
    plt.subplot(1,5,i+1)
    plt.imshow(img)
    plt.axis('off')
plt.show()
```
