```python
#Checking for miising value,duplicate data,incorrect data and perform data cleaning
data.isnull().values.any()
```

```python
data.isna().sum()
```

```python
#Replace NaN value with meadian value---median imputation
for i in data.columns[:17]:
    median_value=data[i].median()
    data[i]=data[i].fillna(median_value)
```

```python
data.isna().sum()
```

```python
data.boxplot(figsize=(35,50))
```

```python
#to treat the outliers
for col_name in data.columns[:-1]: #all column except last one
    q1=data[col_name].quantile(0.25)
    q3=data[col_name].quantile(0.75)
    iqr=q3-q1
    low=q1-1.5*iqr
    high=q3+1.5*iqr
    data.loc[(data[col_name]<low)|(data[col_name]>high),col_name]=data[col_name].median()
```

```python
data['class']=data['class'].astype('category')
```

```python
#we have to encode the categorical data
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
data['class']=labelencoder.fit_transform(data['class'])
```

```python
data.corr()
```

```python
x=data.iloc[:,0:18]
y=data.iloc[:,18]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
from sklearn import metrics
print("Accuracy of Logistic Regression Model on training data",model.score(x_train,y_train))
print("Accuracy of Logistic Regression Model on training data",model.score(x_test,y_test))
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(x_train,y_train)
pred=model.predict(x_test)
print("Accuracy of Naive Bayes in training data",model.score(x_train,y_train))
print("Accuracy of Naive Bayes in test data",model.score(x_test,y_test))
```

```python
from sklearn.svm import SVC
model=SVC()
model.fit(x_train,y_train)
pred=model.predict(x_test)
print("Accuracy of SVM on training data=",model.score(x_train,y_train))
print("Accuracy of Naive Bayes in test data",model.score(x_test,y_test))
```

```python
from scipy.stats import zscore
```

```python
#Standardize the data
x_z=x.apply(zscore)
```

```python
#Finding the patterns
cov_matrix=np.cov(x_z.T)   #T means transpose
```

```python
eigenvalues,eigenvectors=np.linalg.eig(cov_matrix) #linalg=linear algebra
```

```python
eigen_pairs=[(np.abs(eigenvalues[i]),eigenvectors[:,i]) for i in range(len(eigenvalues))]                     #abs=absolute
```

```python
eigen_pairs.sort(reverse=True)
```

```python
tot=sum(eigenvalues)
var_exp=[(i/tot)*100 for i in sorted(eigenvalues,reverse=True)] #each data divideded by total and multiply by 100
cum_var_exp=np.cumsum(var_exp)
print("Cumulative Variance Explained",cum_var_exp)
plt.plot(var_exp)
plt.show()
```
