```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from scipy.stats import zscore
from sklearn.neighbors import KNeighborsClassifier
```

```python
data=pd.read_csv('wisc_bc_data.csv')
```

```python
data.drop('id',axis=1,inplace=True)
```

```python
x=data.drop('diagnosis',axis=1)
y=data['diagnosis']
```

```python
#applying Z_Score on to the dataframe
#how to apply external function
#=Using apply function
x_scaled=x.apply(zscore)#we apllied scaled function because we want thr range of value b/n -1 to 1
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.30,random_state=1)
```

```python
model=KNeighborsClassifier(n_neighbors=5,weights='distance')
```

```python
model.fit(x_train,y_train)
```

```python
predict=model.predict(x_test)
```

```python
from sklearn import metrics
print("Confusion matrix")
c=metrics.confusion_matrix(y_test,predict,labels=["M","B"])
```
